```python
import json
from datetime import datetime
import polars as pl
```

```python
tower = json.loads(open("../dataset/jp-tower.json", "r").read())
```

```python
col_list = [
    "id",
    "owner",
    "datetaken",
    "title",
    "tags",
    "latitude",
    "longitude",
    "place_id",
]

df_towers = []
for k, v in tower["geoJson"]["bulky"].items():
    # ["tower"]["features"][0]
    for item in v["features"]:
        body = {"category": k}
        for ik, iv in item["properties"].items():
            if ik in col_list:
                if ik == "datetaken":
                    body[ik] = int(datetime.strptime(iv, "%Y-%m-%d %H:%M:%S").timestamp())
                else:
                    body[ik] = iv
        body["img_id"] = "-".join([
            k, str(body.get("latitude")), str(body.get("longitude")), str(body.get("datetaken")),
            body.get("owner"), body.get("id"), "n"
        ]) + ".jpg"

        df_towers.append(body)
    break
```

```python
df_towers = pl.DataFrame(df_towers)
df_towers.head()
```

```python
df_towers.write_csv("../dataset/csv/towers.csv")
```
